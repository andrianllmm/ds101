# Notes

## Relational Databases

- SQL is used to create, query, and manage relational databases.
- Tables hold data
- rows = records, columns = fields.
- Each table should have a unique ID (primary key).
- Table & field naming should be consistent.
- Data types: INT, NUMERIC, VARCHAR, BOOLEAN.

## Basic SQL Queries

Basic query structure:

```sql
SELECT column1, column2, ...
FROM table_name
```

Avoid duplicate rows:

```sql
SELECT DISTINCT genre FROM books;
```

Give tables or columns temporary names for readability using aliasing:

```sql
SELECT title AS book_title FROM books;
```

Save query results for reuse and sharing:

```sql
CREATE VIEW recent_books AS
SELECT * FROM books WHERE publication_year > 2020;
```

Restrict results to a set number of rows.

```sql
SELECT * FROM books
LIMIT 5;
```

# Application

Retrieve a list of unique book titles published after 2020, showing only the first 5 results.

```sql
-- Select unique book titles published after 2020
SELECT DISTINCT title AS book_title
FROM books
WHERE publication_year > 2020
LIMIT 5;
```
